```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import time

class StockPortfolio:
    def __init__(self):
        self.stocks = {}

    def add_stock(self, ticker, quantity):
        self.stocks[ticker] = quantity

    def remove_stock(self, ticker):
        if ticker in self.stocks:
            del self.stocks[ticker]

    def track_performance(self):
        portfolio_value = 0
        for ticker, quantity in self.stocks.items():
            stock_data = yf.Ticker(ticker)

            # Print the available keys to see what data is returned
            print(f"Available keys for {ticker}: {stock_data.info.keys()}")

            if 'regularMarketPrice' in stock_data.info:
                current_price = stock_data.info['regularMarketPrice']
            elif 'currentPrice' in stock_data.info:
                current_price = stock_data.info['currentPrice']
            elif 'previousClose' in stock_data.info:
                current_price = stock_data.info['previousClose']
            else:
                print(f"Error: No suitable price key found for {ticker}")
                continue  # Skip this stock if no price is found

            portfolio_value += current_price * quantity
        return portfolio_value

def track_portfolio_performance(portfolio):
    portfolio_values = []
    for i in range(10):  # Track for 10 iterations
        portfolio_values.append(portfolio.track_performance())
        plt.plot(portfolio_values)
        plt.xlabel('Time')
        plt.ylabel('Portfolio Value')
        plt.title('Portfolio Performance Over Time')
        plt.show()

def main():
    portfolio = StockPortfolio()
    portfolio.add_stock('MSFT', 100)
    portfolio.add_stock('AAPL', 50)

    # Run the function once immediately
    track_portfolio_performance(portfolio)

if __name__ == "__main__":
    main()
```
